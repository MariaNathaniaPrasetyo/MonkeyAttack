### 1. Import Library

Bagian ini berfungsi untuk mengimpor semua library Python yang dibutuhkan untuk analisis data. Ini termasuk `pandas` untuk manipulasi data, `folium` untuk visualisasi peta, `sklearn.cluster.KMeans` untuk algoritma clustering, serta `matplotlib.pyplot` dan `seaborn` untuk plotting grafik dan visualisasi data lainnya. Mengimpor library di awal memastikan semua fungsionalitas yang diperlukan tersedia sebelum digunakan.

```python
import pandas as pd
import folium
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

print("Library berhasil di-load, Tuan Gigha!")
```

### 2. Load & Cleaning Data

Pada bagian ini, data mentah dimuat dari file CSV yang diharapkan diunduh dari GBIF. Data kemudian dibersihkan dengan hanya memilih kolom lintang (`decimalLatitude`) dan bujur (`decimalLongitude`), serta menghapus baris yang memiliki nilai kosong (`dropna()`) untuk memastikan kualitas data. Untuk mencegah _overload_ memori di Google Colab, data dibatasi hingga 5000 baris secara acak jika dataset terlalu besar. Ini adalah langkah penting untuk menyiapkan data sebelum analisis lebih lanjut.

```python
file_path = 'DataMonyet.csv'
df = pd.read_csv(file_path, sep='\t', on_bad_lines='skip', low_memory=False)

df_clean = df[['decimalLatitude', 'decimalLongitude']].dropna()

if len(df_clean) > 5000:
    df_clean = df_clean.sample(5000, random_state=42)

print(f"Total titik koordinat yang siap diproses: {len(df_clean)}")
```

### 3. Machine Learning: K-Means Clustering

Bagian ini menerapkan algoritma _clustering_ K-Means untuk mengelompokkan titik-titik koordinat geografis. Dalam contoh ini, data monyet diasumsikan akan dibagi menjadi 5 wilayah (`k = 5`) atau klaster utama. K-Means bekerja dengan mengidentifikasi pusat klaster (centroid) dan mengelompokkan titik-titik data terdekat ke centroid tersebut. Ini membantu dalam mengidentifikasi 'hotspot' atau area konsentrasi serangan monyet.

```python
k = 5
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df_clean['Cluster'] = kmeans.fit_predict(df_clean[['decimalLatitude', 'decimalLongitude']])
```

### 4. Visualisasi: Pemetaan Hotspot Monkey Attack (Revisi)

Setelah proses _clustering_ selesai, bagian ini akan memvisualisasikan hasil klasterisasi pada peta interaktif menggunakan library `folium`. Peta diatur untuk berpusat di Indonesia (`map_center`) dan setiap titik data akan ditampilkan sebagai `CircleMarker` dengan warna yang berbeda, sesuai dengan klaster tempatnya berada. Pada revisi ini, telah ditambahkan konversi eksplisit ke tipe data bilangan bulat (`int()`) untuk `cluster_id` dan bilangan desimal (`float()`) untuk `decimalLatitude` serta `decimalLongitude` untuk memastikan keandalan saat memplot data. Visualisasi ini memudahkan untuk mengidentifikasi dan memahami persebaran geografis dari klaster 'monkey attack'.

```python
# 4. VISUALISASI: PEMETAAN HOTSPOT MONKEY ATTACK (REVISI)
# Bikin peta, kita set posisinya di tengah-tengah Indonesia
map_center = [-2.5489, 118.0149] # Koordinat tengah Indonesia
m = folium.Map(location=map_center, zoom_start=5, tiles='CartoDB positron')

# Bikin warna untuk masing-masing klaster
colors = ['red', 'blue', 'green', 'purple', 'orange']

# Masukkan titik-titik monyet ke dalam peta
for idx, row in df_clean.iterrows():
    # Kita tambahkan int() untuk cluster dan float() untuk koordinat biar aman
    cluster_id = int(row['Cluster'])

    folium.CircleMarker(
        location=[float(row['decimalLatitude']), float(row['decimalLongitude'])],
        radius=3,
        color=colors[cluster_id],
        fill=True,
        fill_color=colors[cluster_id],
        fill_opacity=0.7,
        popup=f"Cluster {cluster_id}"
    ).add_to(m)

# Tampilkan Peta
print("Visualisasi MONKEY ATTACK berhasil dibuat!")
m
```
